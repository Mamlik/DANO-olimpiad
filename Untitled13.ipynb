```python
data = pd.read_csv("/content/gdrive/MyDrive/DANO/tinkoff_chashback.csv")
data
```

```python
# смотрим на отрицательный кешбек, удаляем такие запасы
data = data[(data['cb_merch_last_3_month'] >= 0) & (data['cb_bank_last_3_month'] >= 0) & (data['cb_merch_before'] >= 0) & (data['cb_merch_before'] >= 0)]
data
```

```python
# удаляем колонки: [client_id, Taxi, Taxi_before]
data.drop(axis=0, columns = ["Taxi", "Taxi_before", "client_id", 'cb_bank_last_3_month', 'cb_bank_before'], inplace = True)
data
```

```python
# нормальным значением

emission = dict()
for cur in data:
    emission[cur] = np.percentile([cur for cur in data[cur] if cur != 0], [1, 99])
    print(cur, emission[cur])
emission['cnt_communication'] = [0, 11]  # устанавливае диапозон отдельный для "cnt_communication"
```

```python
# ставим последовательные индексы
data.index = [cur for cur in range(len(data))]
drop_list = list()
for idx in range(len(data)):
    cur = data.iloc[idx]
    counter = 0
    for key in data:
        if cur[key] != 0 and (cur[key] < emission[key][0] or cur[key] > emission[key][1]):
            counter += 1
    if counter > 3:
        drop_list.append(idx)
```

```python
print(len(drop_list))
print(len(data))
data.drop(labels=drop_list, axis=0, inplace=True)
data.index = [cur for cur in range(len(data))]
```

```python
# удаляем записи в которых количество активированных предложений превосходит колиличество просмотренных
data = data[data['cnt_activation'] < data['cnt_view']]
```

**Добавляем НОВЫЕ признаки**
1. Среднемесяцный кеш за год - year_cesk
2. Среднемесячный банковский кеш за год - year_cb_bank
3. Среднемесячный партнёрский кеш за год - year_cb_merch
4. Среднемесячный сумма трат за год - year_spend
5. Среднемесячный сумма трат за 3 месяца - spend_3_month
6. Среднемесячный сумма трат за 8 месяцев - spend_before
7. Среднемесячный общий кеш за 3 месяца - cach_3_month
8. Среднемесячный общий кеш за 8 месяцев - cash_before
9. cnt_view / cnt_activation

```python
data['year_spend'] = data[data.columns[5:]].sum(axis=1) / 2
data['spend_3_month'] = data[data.columns[5:21]].sum(axis=1)
data['spend_before'] = data[data.columns[23:-3]].sum(axis=1)
data['year_cash'] = data['cb_merch_before'] + data['cb_merch_last_3_month']
data['coff'] = data['cnt_view'] / data['cnt_activation']
data
```

```python
'''
Удаляем записи с аномально маленьким кешбеком
'''
data = data[data['year_cash'] >= 0]
```

```python
# удаляем клиентов с аномально маленькими общими тратами
data = data[data['year_spend'] > 100]
```

```python
print(
    data['year_spend'].mean(), '\n',
    min(data['year_spend']), '\n',
    max(data['year_spend'])
)
```

```python
# Удалим клиентов кооторые перестали пользоваться банком(year_spend == 0)
# КАК С НУЛЕВЫМИ ТРАТАМИ У ПОЛЬЗОВАТЕЛЯ МОЖЕТ БЫТЬ КЕШ ??????? таких 726 записей
data = data[data['year_spend'] != 0]
```

**Делим данные на 4 класса**

3. Клиенты которые не использовали кешбек первые 8 месяцев, а поотом решили использовать.
4. Клиеты которые игнорировали кешбек.

```python
# считаем средние траты для за период пользование кешеком
class_1_data = data[(data["cb_merch_before"] != 0) & (data["cb_merch_last_3_month"] != 0)] # 1 1
class_2_data = data[(data["cb_merch_before"] != 0) & (data["cb_merch_last_3_month"] == 0)] # 1 0
class_3_data = data[(data["cb_merch_before"] != 0) & (data["cb_merch_last_3_month"] == 0)] # 0 1
class_4_data = data[(data["cb_merch_before"] != 0) & (data["cb_merch_last_3_month"] == 0)] # 0 0
```
